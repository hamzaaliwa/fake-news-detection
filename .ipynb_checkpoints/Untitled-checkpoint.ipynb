```python

def preprocess_text(text):    
    # Retirer la ponctuation et les caractères spéciaux
    cleaned_text =re.sub(r'[^\w\s]', '', text)
    
    # Normaliser le texte en minuscules
    cleaned_text = cleaned_text.lower()
    
    # Tokenisation des mots
    tokens = word_tokenize(cleaned_text)
    
    # Retirer les stopwords
    stop_words = set(stopwords.words("english"))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    
    # Lemmatisation
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]
    
    return lemmatized_tokens

# Appliquer le prétraitement à chaque ligne dans le DataFrame
#BD1
datafinal['newtext'] = datafinal['text'].apply(preprocess_text)
datafinal['newtitle'] = datafinal['title'].apply(preprocess_text)
datafinal['newauthor'] = datafinal['author'].apply(preprocess_text)
print("good")
```

```python
# POUR BD2

from gensim import corpora, models

# Tokenize your text data and create a dictionary
tokenized_text = [text.split() for text in datafinal2['text']]
dictionary = corpora.Dictionary(tokenized_text)

# Create a corpus
corpus = [dictionary.doc2bow(text) for text in tokenized_text]

# Train the LDA model
lda_model = models.LdaModel(corpus, num_topics=5, id2word=dictionary, passes=15)

```

```python
from wordcloud import WordCloud
# Create word clouds for each topic
for topic_id in range(lda_model.num_topics):
    words = lda_model.show_topic(topic_id, topn=10)
    topic_words = ' '.join([word[0] for word in words])
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(topic_words)
    
    plt.figure(figsize=(8, 4))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Topic {topic_id}')
    plt.show()
```

```python

# Get the topic distribution for each text
topic_distribution = [lda_model.get_document_topics(text) for text in corpus]

# Extract the dominant topic for each text
dominant_topics = [max(topics, key=lambda x: x[1])[0] for topics in topic_distribution]

# Add the dominant topic information to your DataFrame
datafinal2['dominant_topic'] = dominant_topics

# Create a bar chart to visualize topic distribution by class
topic_counts = datafinal2.groupby(['label', 'dominant_topic']).size().unstack(fill_value=0)
topic_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Topic Distribution by Class')
plt.legend(title='Dominant Topic')
plt.show()
```
